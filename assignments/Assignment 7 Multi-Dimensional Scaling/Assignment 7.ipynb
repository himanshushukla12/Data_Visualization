```python
import geopy 
from geopy.distance import distance
```

```python
# Function to compute distance between two (latitude, longitude) pairs
def compute_distance(latlon1, latlon2):
    return distance(latlon1, latlon2).km

# Compute distance matrix
distance_matrix = {}
for i in range(len(df)):
    city1 = df['city'][i]
    latlon1 = (df['lat'][i], df['lng'][i])
    distance_matrix[city1]={}
    for j in range(len(df)):
        city2 = df['city'][j]
        latlon2 = (df['lat'][j], df['lng'][j])
        dist = compute_distance(latlon1, latlon2)
        distance_matrix[city1][city2]= dist
        print(f'Distance from {city1} to {city2} is {dist}')
    
```

```python
distance_matrix
```

```python
# Create a zone column based on admin_name
zones = {
    "North": ['Delhi', 'Himāchal Pradesh', 'Punjab', 'Uttar Pradesh', 'Haryāna', 'Jammu and Kashmīr', 'Chandīgarh'],
    "East": ['Bihār', 'Odisha', 'Jhārkhand', 'West Bengal'],
    "West": ['Rājasthān' , 'Gujarāt', 'Goa', 'Mahārāshtra'],
    "South": ['Andhra Pradesh', 'Telangāna', 'Karnātaka', 'Kerala', 'Tamil Nādu', 'Puducherry'],
    "Central": ['Madhya Pradesh', 'Chhattīsgarh'],
    "North East": ['Assam']
}

def assign_zone(city_name):
    for zone, cities in zones.items():
        if city_name in cities:
            return zone
    return None

df['zone'] = df['admin_name'].apply(assign_zone)
df
```

```python
from sklearn.manifold import MDS
# Extract city names and distances from the dictionary
cities = list(distance_matrix.keys())
distances = np.array([[distance_matrix[city1][city2] for city2 in cities] for city1 in cities])

# Use MDS for 2D representation
mds = MDS(random_state=0, dissimilarity='precomputed', n_components=2)
fit_data = mds.fit_transform(distances)

# fit_data now contains the 2D coordinates for the cities based on their distances
print(fit_data)
```
